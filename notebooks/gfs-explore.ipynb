```python
from datetime import datetime, timedelta
from herbie import FastHerbie
```

```python
t0 = datetime.utcnow().replace(hour=0, minute=0, second=0, microsecond=0) - timedelta(days=1)

fh = FastHerbie(
    [t0],
    model="gfs",
    fxx=range(0, 6),
    prod="pgrb2.0p50",
    priority=["google", "aws", "nomads", "azure"],
)
fh
```
